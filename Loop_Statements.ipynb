For Loop Statement

```python
week= ["Sunday", "Monday", "Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
 print(x)
```

```python
week= ["Sunday", "Monday", "Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  print(x)
  if x == "Thursday":
    break
```

```python
#The range() Function
for x in range(6):
  print(x)

for x in range (2,5):
  print(x)
```

Nested Loops

```python
adjective = ["red","big","tasty"]
fruits = ["apple","banana","cherry"]

for x in adjective:
  for y in fruits:
   print(x,y)
```

Phyton While Loops

```python
i = 1
while i < 6:
  print(i)
  i += 1 #Assignment Operator
```

The Continue Statement

```python
i=0
while i<6:
  i+=1
  if i == 3:
    continue
  if i== 6:
    break
  print(i)
```

The Else Statement

```python
i =1
while i<6:
  print(i)
  i+= 1
else:
  print("i is no longer than 6")
```
